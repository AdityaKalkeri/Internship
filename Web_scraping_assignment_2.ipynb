# Task 1: naukri.com: Data Analyst

```python
def data_analyst():
    driver = webdriver.Firefox(executable_path = (GeckoDriverManager().install()))
    csv_file = open('data_analyst.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Job','Company','experience', 'salary', 'location', 'description'])
    driver.get('https://www.naukri.com/')
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="qsb-keyword-sugg"]').send_keys('Data Analyst')
    driver.find_element_by_xpath('//*[@id="qsb-location-sugg"]').send_keys('Bangalore')
    driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[2]/section/div/form/div[3]/button').click()
    time.sleep(3)
    driver.execute_script("window.open('');")
    article=driver.find_elements_by_tag_name('article')
    for i in article:
        url = i.find_element_by_tag_name('a').get_attribute('href')
        try:
            driver.switch_to.window(driver.window_handles[1])
            driver.get(url)
            time.sleep(5)
        except:
            print('get Url problem')
        try:
            print(driver.find_element_by_class_name('jd-header-title').text)
            title = driver.find_element_by_class_name('jd-header-title').text
        except:
            print(driver.find_element_by_class_name('av-special-heading-tag ').text)
            title = driver.find_element_by_class_name('av-special-heading-tag ').text
        try:
            print(driver.find_element_by_class_name("pad-rt-8").text)
            company = driver.find_element_by_class_name("pad-rt-8").text
        except:
            print(driver.find_element_by_class_name('logo').find_element_by_tag_name('img').get_attribute('alt'))
            company = driver.find_element_by_class_name('logo').find_element_by_tag_name('img').get_attribute('alt')
        try:
            print(driver.find_element_by_class_name("exp").text)
            exp = driver.find_element_by_class_name("exp").text
        except:
            print(driver.find_element_by_class_name('slide-meta').text)
            exp = driver.find_element_by_class_name('slide-meta').text
        try:
            print(driver.find_element_by_class_name("salary").text)
            salary  = driver.find_element_by_class_name("salary").text
        except:
            print(driver.find_elements_by_class_name('slide-meta')[2].text)
            salary = driver.find_elements_by_class_name('slide-meta')[2].text
        try:
            print(driver.find_element_by_class_name("location").text)
            location = driver.find_element_by_class_name("location").text
        except:
            print(driver.find_elements_by_class_name('slide-meta')[1].text)
            location = driver.find_elements_by_class_name('slide-meta')[1].text
        try:
            print(driver.find_element_by_class_name('job-desc').text.replace('\n',' '))
            desc = driver.find_element_by_class_name('job-desc').text.replace('\n',' ')
        except:
            print(driver.find_element_by_tag_name("section").text.replace('\n', ' '))
            desc = driver.find_element_by_tag_name("section").text.replace('\n', ' ')
        writer.writerow([title,company,exp, salary, location, desc])
        print()
        driver.switch_to.window(driver.window_handles[0])
    csv_file.close()
```
