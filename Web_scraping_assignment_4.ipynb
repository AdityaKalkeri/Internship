```python
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
import time
import csv
import pandas as pd
```

# 1) Wikipedia: Youtube video with most views

```python
def wiki():
    csv_file = open('youtube_rankings.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Rank', 'Name', 'Uploader/Artist', 'Views', 'Date'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
    time.sleep(5)
    for i in range(1,31):
        rank = i
        try:
            name = driver.find_element_by_css_selector('table.wikitable:nth-child(20) > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(2) > a:nth-child(1)'.format(i)).text
        except:
            body = driver.find_elements_by_css_selector('table.wikitable:nth-child(20) > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(2) > i:nth-child(1) > a:nth-child(1)'.format(i))
            if len(body) > 0:
                name = body[0].text
            else:
                name = driver.find_element_by_css_selector('table.wikitable:nth-child(20) > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(2)'.format(i)).text
        artist = driver.find_element_by_css_selector('table.wikitable:nth-child(20) > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(3)'.format(i)).text
        views = driver.find_element_by_css_selector('table.wikitable:nth-child(20) > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(4)'.format(i)).text
        date = driver.find_element_by_css_selector('table.wikitable:nth-child(20) > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(5)'.format(i)).text
        writer.writerow([i, name, artist, views, date])
    csv_file.close()
```

```python
wiki()
pd.read_csv('youtube_rankings.csv')
```

# 2) BCCI: India match fixtures

```python
def bcci():
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    csv_file = open('matches.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Date', 'Time', 'Format', 'Series', 'Location'])
    driver.get('https://www.bcci.tv/')
    time.sleep(5)
    url = driver.find_element_by_css_selector('.navigation__list > li:nth-child(1) > div:nth-child(2) > div:nth-child(1) > ul:nth-child(2) > li:nth-child(1) > a:nth-child(1)').get_attribute('href')
    driver.get(url)
    time.sleep(5)
    for i in range(1,25):
        try:
            date = driver.find_element_by_css_selector('a.js-match:nth-child({}) > div:nth-child(1)'.format(i)).text.split('\n')
            tme = ' '.join([str(elem) for elem in date[3:]])
            date = ' '.join([str(elem) for elem in date[1:3]])

            match_type = driver.find_element_by_css_selector('a.js-match:nth-child({}) > div:nth-child(2) > div:nth-child(2) > p:nth-child(2) > strong:nth-child(1)'.format(i)).text

            series = driver.find_element_by_css_selector('a.js-match:nth-child({}) > div:nth-child(2) > div:nth-child(1) > span:nth-child(2)'.format(i)).text

            location = driver.find_element_by_css_selector('a.js-match:nth-child({}) > div:nth-child(2) > div:nth-child(2) > p:nth-child(2) > span:nth-child(2)'.format(i)).text

            writer.writerow([date,tme,match_type,series,location])
        except:
            pass
        
    csv_file.close()
```

```python
bcci()
pd.read_csv('matches.csv')
```

```python
def guru99():
    csv_file = open('exception.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Exception', 'Description'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://www.guru99.com')
    driver.find_element_by_css_selector('.cloumnsize1 > ul:nth-child(2) > li:nth-child(3) > a:nth-child(1)').click()
    driver.find_element_by_css_selector('table.table:nth-child(25) > tbody:nth-child(1) > tr:nth-child(34) > td:nth-child(1) > a:nth-child(2)').click()
    for i in range(2,43):
        exception = driver.find_element_by_css_selector('.table > tbody:nth-child(1) > tr:nth-child({}) > td:nth-child(1)'.format(i)).text
        desc = driver.find_element_by_css_selector('.table > tbody:nth-child(1) > tr:nth-child({}) > td:nth-child(2)'.format(i)).text
        writer.writerow([exception, desc])
    csv_file.close()
```

# 4) Statistictime: Indian State GDP

```python
def stategdp():
    csv_file = open('gdp.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Rank','State', 'GDP_2020', 'GDP_2019', 'GDP_billions', 'GDP_share'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('http://statisticstimes.com/')
    time.sleep(5)
    url = driver.find_element_by_css_selector('div.dropdown:nth-child(2) > div:nth-child(2) > a:nth-child(3)').get_attribute('href')
    driver.get(url)
    time.sleep(5)
    driver.find_element_by_css_selector('body > div:nth-child(2) > div:nth-child(6) > div:nth-child(2) > ul:nth-child(2) > li:nth-child(1) > a:nth-child(1)').click()
    time.sleep(5)
    rank = 1
    for i in range(1,34):
        try:
            state = driver.find_element_by_css_selector('#table_id > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(2)'.format(i)).text
            gdp_20= driver.find_element_by_css_selector('#table_id > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(3)'.format(i)).text
            gdp_19=driver.find_element_by_css_selector('#table_id > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(4)'.format(i)).text
            gdp_billions=driver.find_element_by_css_selector('#table_id > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(6)'.format(i)).text
            share = driver.find_element_by_css_selector('#table_id > tbody:nth-child(3) > tr:nth-child({}) > td:nth-child(5)'.format(i)).text
            writer.writerow([rank, state, gdp_20, gdp_19, gdp_billions, share])
        except:                                    
            continue
    csv_file.close()
```

```python
stategdp()
pd.read_csv('gdp.csv')
```

# 5) Github: Repositories

```python
def github():
    csv_file = open('github.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Title', 'Language', 'No. of Contributors', 'Description'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://github.com/')
    time.sleep(5)
    url = driver.find_element_by_css_selector('ul.mb-3:nth-child(3) > li:nth-child(3) > a:nth-child(1)').get_attribute('href')
    driver.get(url)
    time.sleep(5)
    for i in range(1,26):
        try:
            title = driver.find_element_by_css_selector('article.Box-row:nth-child({}) > h1:nth-child(2) > a:nth-child(1)'.format(i)).text
        except:
            title = '-'
        try:
            desc = driver.find_element_by_css_selector('article.Box-row:nth-child({}) > p:nth-child(3)'.format(i)).text
        except:
            desc = '-'
        try:    
            lang = driver.find_element_by_css_selector('article.Box-row:nth-child({}) > div:nth-child(4) > span:nth-child(1) > span:nth-child(2)'.format(i)).text
        except:
            lang = '-'
        try:
            contri = driver.find_element_by_css_selector('article.Box-row:nth-child({}) > div:nth-child(4) > span:nth-child(4)'.format(i))
            num = len(contri.find_elements_by_class_name('d-inline-block'))
        except:
            num = '-'
        writer.writerow([title, lang, num, desc])
    csv_file.close()
```

```python
github()
pd.read_csv('github.csv')
```

# 6) BillBoard: top100 songs

```python
def billboard():
    csv_file = open('top100.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['Rank', 'Song', 'Artist', 'Last Week Rank', 'Peak Rank', 'Weeks on Chart'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://www.billboard.com/')
    time.sleep(5)
    driver.find_element_by_css_selector('li.header__subnav__item:nth-child(3) > a:nth-child(1)').click()
    time.sleep(10)
    rank = 1
    for i in range(1, 110):
        try:
            try:
                song = driver.find_element_by_css_selector('li.chart-list__element:nth-child({}) > button:nth-child(1) > span:nth-child(2) > span:nth-child(1)'.format(i)).text
            except:
                continue
            try:
                artist = driver.find_element_by_css_selector('li.chart-list__element:nth-child({}) > button:nth-child(1) > span:nth-child(2) > span:nth-child(2)'.format(i)).text
            except:
                artist = '-'
            try:
                last_week = driver.find_element_by_css_selector('li.chart-list__element:nth-child({}) > button:nth-child(1) > div:nth-child(3) > div:nth-child(2)'.format(i)).text
            except:
                last_week = '-'
            try:
                peak = driver.find_element_by_css_selector('li.chart-list__element:nth-child({}) > button:nth-child(1) > div:nth-child(3) > div:nth-child(3)'.format(i)).text
            except:
                peak = '-'
            try:
                weeks = driver.find_element_by_css_selector('li.chart-list__element:nth-child({}) > button:nth-child(1) > div:nth-child(3) > div:nth-child(4)'.format(i)).text
            except:
                weeks = '-'
            writer.writerow([rank, song, artist, last_week, peak, weeks])
            rank = rank+1
        except:
            pass
    csv_file.close()
    
```

```python
billboard()
pd.read_csv('top100.csv')
```

# 8) Guardian: Top selling books

```python
def books():
    csv_file = open('top_books.csv', 'w')
    writer  = csv.writer(csv_file)
    writer.writerow(['Rank', 'Title', 'Author', 'Volume', 'Publisher', 'Genre'])
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
    time.sleep(10)
    rank = 1
    for i in range(0,100):
        title = driver.find_element_by_css_selector('#table-cell-10943-{}-1'.format(i)).text
        author = driver.find_element_by_css_selector('#table-cell-10943-{}-2'.format(i)).text
        volume = driver.find_element_by_css_selector('#table-cell-10943-{}-3'.format(i)).text
        publisher = driver.find_element_by_css_selector('#table-cell-10943-{}-4'.format(i)).text
        genre = driver.find_element_by_css_selector('#table-cell-10943-{}-5'.format(i)).text
        writer.writerow([rank, title, author, volume, publisher, genre])
    csv_file.close()
```
