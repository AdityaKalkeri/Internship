# 1)Amazon

```python
def amazon1():
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://www.amazon.in/')
    time.sleep(3)
    search = input('Amazon Search: Enter Product name')
    driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]').send_keys(search)
    driver.find_element_by_xpath('//*[@id="nav-search-submit-button"]').click()
    time.sleep(3)
    print('Search page is ready')
    return driver, search
```

# 3) Google Images

```python
def googleimg():
    search = input('Enter Search Term Here:')
    driver = webdriver.Firefox(executable_path = GeckoDriverManager().install())
    driver.get('https://images.google.com/')
    time.sleep(5)
    driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/form/div[2]/div[1]/div[1]/div/div[2]/input').send_keys(search)
    driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/form/div[2]/div[1]/div[1]/button/div').click()
    time.sleep(10)
    for i in range(1,100):
        try:
            print('Line 1')
            driver.find_element_by_css_selector('.islrc > div:nth-child({}) > a:nth-child(1) > div:nth-child(1) > img:nth-child(1)'.format(i)).click()
            time.sleep(10)
            print('line 2')
            url = driver.find_element_by_css_selector('div.tvh9oe:nth-child(2) > c-wiz:nth-child(1) > div:nth-child(1) > div:nth-child(1) > div:nth-child(1) > div:nth-child(2) > a:nth-child(1) > img:nth-child(1)').get_attribute('src')
            print('Line 3')
            r = requests.get(url, stream = True)
            print('Requesting image')
            time.sleep(5)
            if r.status_code == 200:
                r.raw.decode_content = True
                with open('image{}'.format(i),'wb') as f:
                    shutil.copyfileobj(r.raw, f)  
                print('Image sucessfully Downloaded: ','image{}'.format(i))
        except:
            print('Image retrieval problem')
```
